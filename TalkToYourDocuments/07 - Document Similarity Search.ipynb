# 🔍 Similarity Search 🔍

In this notebook, we will take the documents that you have loaded setup similarity search.

This will allow you to query on the content of the documents and find similar chunks of text _(based on semantic similarity rather than keyword search)_

> Note: this notebook focuses on semantic similarity search but many hybrid search options are becoming available.  
> See a future notebook for more details.

Quote from GPT-4:
> _"Searching text for keywords, which is more traditional, can be considered a sparse method. This is because documents are typically represented as sparse vectors in a high-dimensional space, where each dimension corresponds to a specific keyword and the value in each dimension corresponds to the frequency of that keyword in the document. Techniques like TF-IDF or BM25 are used in this context, where importance of the words are determined by their frequencies._
> 
> _On the other hand, semantic search is considered a dense method. This is because documents are represented by dense vectors (also known as embeddings) in a lower-dimensional space. These embeddings are typically generated by modern machine learning models (like word2vec, GloVe, BERT, etc.) that are capable of capturing the semantic meaning of the text. The similarity between these dense vectors is calculated using measures like cosine similarity or Euclidean distance, and nearest neighbor search techniques can be used to find the most similar documents to a given query._
>
> _So in summary, keyword search = sparse, semantic search = dense. Keyword search typically uses traditional information retrieval techniques like BM25, while semantic search uses modern machine learning models and vector similarity/distance metrics."_
